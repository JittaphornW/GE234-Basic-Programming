## 🔹 ตัวอย่างที่ 1: การใช้ `for loop` เพื่อแสดงค่าพิกัด


```python

# พิกัดของสถานที่สำคัญในประเทศไทย
locations = {
    "Bangkok": (13.7563, 100.5018),
    "Chiang Mai": (18.7883, 98.9853),
    "Phuket": (7.8804, 98.3923),
    "Ubon Ratchathani": (15.2287, 104.8564)
}

# วนลูปแสดงค่าพิกัด
for city, coord in locations.items():
    print(f"{city}: Latitude {coord[0]}, Longitude {coord[1]}")
```


## 🔹 ตัวอย่างที่ 2: การใช้ `if-else` กรองข้อมูลพิกัด


```python

for city, coord in locations.items():
    if coord[0] > 16:
        print(f"{city} อยู่ทางภาคเหนือของประเทศไทย")
    else:
        print(f"{city} ไม่ใช่จังหวัดทางภาคเหนือ")
```


## 🔹 ตัวอย่างที่ 3: ใช้ `while loop` นับจำนวนจุดข้อมูล


```python

points = [(13.7563, 100.5018), (18.7883, 98.9853), (7.8804, 98.3923)]
count = 0

while count < len(points):
    print(f"Point {count + 1}: {points[count]}")
    count += 1
```


## 🔹 ตัวอย่างที่ 4: วิเคราะห์ข้อมูล Raster ด้วย `for loop`


```python

import numpy as np

# จำลองข้อมูล NDVI (Normalized Difference Vegetation Index)
ndvi = np.array([
    [0.2, 0.4, 0.5],
    [0.1, 0.3, 0.7],
    [0.6, 0.8, 0.2]
])

# วนลูปตรวจสอบค่าพืชพรรณ
for row in range(ndvi.shape[0]):
    for col in range(ndvi.shape[1]):
        if ndvi[row, col] > 0.3:
            print(f"พิกัด ({row}, {col}) มีพืชพรรณหนาแน่น (NDVI={ndvi[row, col]})")
```


# 📝 **กิจกรรมในแลป**

1. **แบบฝึกหัด 1**: เขียน `for loop` เพื่อแสดงชื่อจังหวัดและพิกัดจากลิสต์ของคุณเอง
2. **แบบฝึกหัด 2**: ใช้ `if-else` เพื่อแยกแยะพื้นที่ที่เป็นภูเขาและพื้นที่ราบจากลิสต์ของพิกัด
3. **แบบฝึกหัด 3**: ใช้ `while loop` เพื่อวนลูปและแสดงค่าละติจูดจากลิสต์จนกว่าจะพบค่าต่ำกว่า 10°
4. **แบบฝึกหัด 4**: ใช้ `for loop` วิเคราะห์ค่า NDVI และหาค่าพื้นที่ที่มีสภาพแห้งแล้ง (NDVI < 0.2)


**แบบฝึกหัดที่** 1 for loop

```python
# ค่าพิกัดจากลิสต์ที่หามาได้
locations = {"Lopburi": (14.7995, 100.6533),"Loei":(17.4860, 101.7223),"Nan":(18.7756, 100.7730),"Songkhla":(7.1756, 100.6143),"Ranong":(9.9528, 98.6084)}
# เครดิตฐานข้อมูลจังหวัด และพิกัด ข้อมูลจาก Google map

for city, coord in locations.items():
  print(f"{city}: Latitude {coord[0]}, Longtitude{coord[1]}")
```

**แบบฝึกหัดที่** 2 if-else

```python
# แบบฝึกหัด 2: ใช้ if-else แยกแยะพื้นที่ภูเขาและที่ราบ
for city, coord in locations.items():
    # สมมติเกณฑ์: ละติจูดมากกว่า 15 องศาเหนือเป็นพื้นที่ภูเขา (ภาคเหนือ/อีสานตอนบน)
    if coord[0] > 15:
        print(f"{city}: เป็นพื้นที่ภูเขา")
    else:
        print(f"{city}: เป็นพื้นที่ราบหรือชายฝั่ง")
```

**แบบฝึกหัดที่** 3 while loop

```python
# แปลงข้อมูลจาก dictionary เป็น list ของ tuple (ชื่อเมือง, (lat, lon))
# อ้างอิงข้อมูลจากแบบฝึกหัดที่ 1
city_list = list(locations.items())
i = 0

print("เริ่มการตรวจสอบค่าละติจูด")
while i < len(city_list):
    city, coord = city_list[i]
    lat = coord[0]

    # แสดงค่าละติจูดปัจจุบัน
    print(f"Checking {city}: Latitude {lat}")

    # เงื่อนไขหยุด: ถ้าละติจูดต่ำกว่า 10 องศา
    if lat < 10:
        print(f"== หยุดการทำงาน: พบ {city} มีละติจูดต่ำกว่า 10 องศา ({lat}) ==")
        break

    i += 1
```

**แบบฝึกหัดที่** 4 for loop

```python
# วิเคราะห์ค่า NDVI เพื่อหาพื้นที่แห้งแล้ง (NDVI < 0.2)
# อ้างอิงข้อมูลจากตัวอย่างที่ 4
import numpy as np

# ใช้ข้อมูล NDVI จากตัวอย่างที่ 4
print("ผลการวิเคราะห์พื้นที่แห้งแล้ง:")
found_drought = False

for row in range(ndvi.shape[0]):
    for col in range(ndvi.shape[1]):
        value = ndvi[row, col]
        if value < 0.2:
            print(f"พิกัด ({row}, {col}) เป็นพื้นที่แห้งแล้ง (NDVI = {value})")
            found_drought = True

if not found_drought:
    print("ไม่พบพื้นที่แห้งแล้งในข้อมูลชุดนี้")
```
